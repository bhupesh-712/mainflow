```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv("D:/mainflow/householdtask3.csv")
```

```python
print("Initial DataFrame:")
print(dataset.head(10))
```

```python
def plot_bar_charts(df, cols):
    for column in cols:
        if column not in df.columns:
            print(f"Error: '{column}' is not a valid column in the DataFrame.")
            continue

        if df[column].isnull().sum() > 0:
            print(f"Note: '{column}' contains some missing values.")
        
        years = df['year']
        values = df[column]

        plt.figure(figsize=(10, 5))
        plt.bar(years, values, color='skyblue')
        plt.xlabel('Year')
        plt.ylabel(column.replace('_', ' ').capitalize())
        plt.title(f'Yearly {column.replace("_", " ").capitalize()}')
        plt.grid(axis='y')
        plt.show()
```

```python

def plot_line_charts(df, cols):
    for column in cols:
        if column not in df.columns:
            print(f"Error: '{column}' is not a valid column in the DataFrame.")
            continue

        if df[column].isnull().sum() > 0:
            print(f"Note: '{column}' contains some missing values.")
        
        years = df['year']
        values = df[column]

        plt.figure(figsize=(10, 5))
        plt.plot(years, values, marker='o', color='orange', linestyle='-')
        plt.xlabel('Year')
        plt.ylabel(column.replace('_', ' ').capitalize())
        plt.title(f'Yearly Trend of {column.replace("_", " ").capitalize()}')
        plt.grid(True)
        plt.show()
```

```python
columns_list = ['tot_hhs', 'own', 'own_wm', 'own_prop', 'own_wm_prop',
                'prop_hhs', 'age', 'size', 'income', 'expenditure', 'eqv_income',
                'eqv_exp']
```

```python
plot_bar_charts(dataset, columns_list)
```

```python
plot_line_charts(dataset, columns_list)
```
